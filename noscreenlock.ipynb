```python
import time
import pyautogui
```

```python
def press_num_lock():
    pyautogui.press('numlock')

if __name__ == "__main__":
    try:
        while True:
            press_num_lock()
            time.sleep(30)  # sleep for 30 seconds
    except KeyboardInterrupt:
        print("\nExiting the program.")
```
