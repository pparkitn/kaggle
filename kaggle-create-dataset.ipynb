```python
test_csv = pd.read_csv('/kaggle/input/rsna-breast-cancer-detection/train.csv')
test_csv['randNum'] = np.random.randint(1, 10, test_csv.shape[0])
test_csv.head(5)
```

```python
SOURCE_FOLDER = "/kaggle/input/rsna-mammography-images-as-pngs/images_as_pngs_512/train_images_processed_512/"
SOURCE_FOLDER = "/kaggle/input/rsna-mammography-images-as-pngs/images_as_pngs_1024/train_images_processed_1024/"
SAVE_FOLDER = "/tmp/working/data/1024" 
os.makedirs(SAVE_FOLDER+'/test/pos/', exist_ok=True)
os.makedirs(SAVE_FOLDER+'/test/neg/', exist_ok=True)
os.makedirs(SAVE_FOLDER+'/train/pos/', exist_ok=True)
os.makedirs(SAVE_FOLDER+'/train/neg/', exist_ok=True)
```

```python
%%time

for _, row in test_csv.iterrows():
    patient_id = str(row.patient_id)
    image_id = str(row.image_id)
    randNum = row.randNum
    cancer = row.cancer
    image_name = SOURCE_FOLDER + "/" + patient_id + "/" + image_id +'.png'
    
    if randNum >=8:
        if cancer == 1:
            to_loc = '/test/pos/'
        else:
            to_loc = '/test/neg/'
    else:
        if cancer == 1:
            to_loc = '/train/pos/'
        else:
            to_loc = '/train/neg/'
    
    dest_folder = SAVE_FOLDER + to_loc+ patient_id + '_' + image_id+'.png'    
    
    try:
        shutil.copyfile(image_name,dest_folder)
    except:
        print("FAIL")
```

```python
def get_random_string(length):
    letters = string.ascii_lowercase
    result_str = ''.join(random.choice(letters) for i in range(length))
    return result_str
```

```python
def duplicate_folder_files(filelist,dup_copy):

    filelist = glob.glob(filelist)

    for i, f in enumerate(filelist):    
        patient, image = f.split('/')[-1][:-4].split('_')
        new_file_name = f 
        rep = patient+'_'+image+'.png'
                
        for i in range(0,dup_copy,1):
            try:
                prefix = get_random_string(10)
                new_name = new_file_name.replace(rep, prefix+'_'+rep)
                shutil.copyfile(f,new_name)
            except:
                print('FAIL')
```

```python
%%time

duplicate_folder_files(SAVE_FOLDER+'/test/pos/*',47)
duplicate_folder_files(SAVE_FOLDER+'/train/pos/*',44)
```
